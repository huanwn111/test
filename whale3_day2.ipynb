```python
#需求：简单线性回归代码实现
#公式1 y = wx + b 变量x 结果为y，预测使用损失函数最小，即 （实际y值-y预测值）^2 最小
#公式2 w = 求合(x-x均值)(y-y均值)/求合(x-x均值)^2
#公式3 b = y均值 - w*x均值 = y均值/x均值
##### y = b1x + b0
import numpy as np
import time
from matplotlib import pyplot as plt
%matplotlib inline

#==算术法求解==
#预测x为给定值时的y值
x_next = 15.

#回归模型
def simpleLinearRegresion(x,y):#x,y是一组数的列表
    n = len(x)
    upper = 0.#分子
    under = 0.#分母
    
    for i in range(0,n):
        #公式计算w 
        #代入公式2
        #分子
        upper += (x[i]-np.mean(x))*(y[i]-np.mean(y))
        #分母
        under += (x[i]-np.mean(x))**2
        w = float(upper)/float(under)
        #代入公式3
        b = float(np.mean(y))/float(np.mean(x))
        #b = np.mean(y) - w*np.mean(x)
        
        return w,b
    
#预测函数    
def predict(x,b,w):
    #代入公式1
    return w*x + b
    
x = [1.,2.,3.,4.,5.,6.,7.]      
y = [1.,2.,3.,4.,5.,6.,7.]

#验证模型

#线性回归规律模型，让机器学习x y规律
startT = time.time()
b,w = simpleLinearRegresion(x,y)
#全用新x值，预测y值
y_pred = predict(x_next,b,w)
endT = time.time()
print("y_pred %s" % y_pred)
print("回归到预测完成用时：%s"%(endT-startT))

#预测前原数据画图
plt.plot(x,y)
plt.title('line chart before predict')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

#预测后加新x和预测y 画图
x_new = x+[x_next]
y_new = y+[y_pred]

plt.plot(x_new,y_new)
plt.title('line chart after predict')
plt.xlabel('x_new')
plt.ylabel('y_new')
plt.show()
#【遗留问题：精确度有一位数偏差，待调试】
```
