```python
#波士顿房产数据 sklearn库自带的，取用即可
from sklearn.datasets import load_boston
import pandas as pd

bostonDatas = load_boston()
#print(bostonDatas.data)#查看数据是数组格式
#查看下数据
X_features = bostonDatas.data#取特征X
y_labels = bostonDatas.target#取结果值y
X_cols = bostonDatas.feature_names#取特征名称（列名）
#print("特征：%s" % X_features)
#print("目标值（房价）：%s" % y_labels)
print("特征名称：%s " % X_cols)

#转为pandas格式
#特征
bostonDF_X = pd.DataFrame(bostonDatas.data,columns=X_cols)
bostonDF_X.head()
```

```python
#结果
bostonDF_y = pd.DataFrame(bostonDatas.target)
bostonDF_y.head()

#【注意：以下可以不必如此周折，直接boston.data[:,5]取到第5列即是RM,
#fit的时候可以用数组值.reshape(-1,1)】
```

```python
#拼合特征和结果DataFrame
bostonDF = pd.concat([bostonDF_X,bostonDF_y],axis=1)
bostonDF.head()
```

```python
#过滤，只留RM 房间数特征
roomDatas = bostonDF[['RM',0]]
roomDatas.rename(columns={0:'roomPrice'},inplace=True)#房价改标题
roomDatas.head()

#取特征
X_data = roomDatas['RM']
y_data = roomDatas['roomPrice']
print("原始数据特征（房间数）：%s " % X_data)
print("原始数据结果（房价）：%s " % y_data)
```

```python
#数据切分（train_test_split）用sklearn的交叉验证，切分数据为训练集和验证集
from sklearn.cross_validation import train_test_split
#交叉验证，切分训练集和验证集
X_train,X_test,y_train,y_test = train_test_split(X_data,y_data,test_size=0.3,random_state=0)
#切分比例为30%做验证集
#打印长度
print("特征训练集：%s" % len(X_train))
print("特征验证集：%s" % len(X_test))
print("结果训练集：%s" % len(y_train))
print("结果验证集：%s" % len(y_test))
```

```python
#训练模型(用线性回归)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train.values.reshape(-1,1),y_train.values.reshape(-1,1))
#【X_train为series数据，取值变数组reshape进行训练，#不转的话报错，只有一个特征】
```

```python
#预测
y_pred = lr.predict(X_test.values.reshape(-1,1))
```

```python
#评估模型
from sklearn import metrics
import numpy as np

#MSE评估指标 均方误差
skl_MSE = metrics.mean_squared_error(y_test,y_pred)

#RMSE评估指标 均方根误差
skl_RMSE = np.sqrt(skl_MSE)

#算MAE平均绝对值误差 #单词absolute，绝对的
skl_MAE = metrics.mean_absolute_error(y_test,y_pred)
#算R^2
skl_R2 = metrics.r2_score(y_test,y_pred)

print("skl算MSE: %s" % skl_MSE)
print("skl算RMSE: %s" % skl_RMSE)
print("skl算R方: %s" % skl_R2)
print("skl算MAE: %s" % skl_MAE)

#公式与解释参考：
#https://www.jianshu.com/p/9ee85fdad150
```

```python
#手算R方，MSE
hand_MSE=np.sum(np.power((y_test.values.reshape(-1,1) - y_pred),2))/len(y_test.values)
#手算RMSE
hand_RMSE = np.sqrt(hand_MSE)
#均方误差公式：y真实值与预测值之差（误差）平方（方）和再求平均（均）
R2=1-hand_MSE/np.var(y_test.values)#代入【R方公式】,R方越大表示右半部损失越小，表示模型越好
#算MAE平均绝对值误差
hand_MAE=np.sum(np.abs((y_test.values.reshape(-1,1) - y_pred)))/len(y_test.values)

print("手算mse: %s " % hand_MSE)
print("手算rmse: %s " % hand_RMSE)
print("手算R方: %s " % R2)
print("手算MAE: %s " % hand_MAE)
```
