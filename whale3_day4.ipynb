```python
#波士顿房产数据 sklearn库自带的，取用即可
from sklearn.datasets import load_boston
bostonDatas = load_boston()
print(bostonDatas.DESCR)#DESCR describe描述，看下数据特征介绍等
```

```python
#把数据转成DataFrame查看对比

import pandas as pd
import numpy as np
#特征转DataFrame格式
bostonDf_X = pd.DataFrame(bostonDatas.data,columns = bostonDatas.feature_names)
bostonDf_X.head()
```

```python
#结果转DataFrame
bostonDf_y = pd.DataFrame(bostonDatas.target,columns=['housePrice'])#【注意加列名要加在[]中】
bostonDf_y.head()
```

```python
#合并DataFrame
bostonDf = pd.concat([bostonDf_X,bostonDf_y],axis =1)#axis =1 为横向操作，即横向合并
bostonDf.shape#506,14
bostonDf.head()
```

```python
#由上可数据可知，多个特征的值范围差距较大，有在0以下的，有在几百的，
#这样在多变量线性回归中，不利于数据训练
#所以在建模回归之前要对数据进行标准化，整合到差不多大小区间，即做标准化处理
```

```python
#使用sklearn preprocessing模块 StandardScaler类
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

bostonStd_1 = scaler.fit_transform(bostonDf.values)
#scale标准化是把数据归化到正态分布，即，均值为0（去均值的中心化），方差为1（方差的规模化）

bostonStd_1.shape#506,14
showbostonStd_1 = pd.DataFrame(bostonStd_1, columns = bostonDf.columns)
showbostonStd_1.head()
#打印数据对比可知标准化后的数据值范围都在一个小范围内了（scale标准化后的数据是达到符合正态分布的归一化）

#【更多sklearn的归化函数对比说明参考：
#https://blog.csdn.net/weixin_40807247/article/details/82793220
#注意使用StandardScaler()较好，标准化优化器scaler可应用到新数据
#】
```

```python
#使用sklearn preprocessing模块 StandardScaler类
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
bostonStd_1 = scaler.fit_transform(bostonDf.values)
#scale标准化是把数据归化到正态分布，即，均值为0（去均值的中心化），方差为1（方差的规模化）

bostonStd_1.shape#506,14
showbostonStd_1 = pd.DataFrame(bostonStd_1, columns = bostonDf.columns)
showbostonStd_1.head()
```

```python
#下面看下不用sklearn，用手写公式实现正态分布的归一化，对比看下
#正态分布公式 x = (x - 均值) / 标准差，即，x = (x - np.average(x)) / np.std(x);
import numpy as np

def Z_ScoreStandard(x,avg,std):
    x = (x - avg) / std
    return x
bV = bostonDf.values
bostonStd_2 =Z_ScoreStandard(bV,np.average(bV),np.std(bV))
#转成DataFrame格式看一下
showbostonStd_2 = pd.DataFrame(bostonStd_2,columns=bostonDf.columns)
showbostonStd_2.head()

#【更多手写（python计算公式实现）归一化方法说明参考：
#https://blog.csdn.net/kryolith/article/details/39770187】

#【遗留问题：手写和sklearn处理的标准化数值有一些差别，没有完全一致，待查】
```

```python
#切分特征和结果为训练集和测试集
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test =\
    train_test_split(bostonStd_1[:,[0,13]],bostonStd_1[:,[13]], test_size = 0.3, random_state=0)
#分别切分X（标准化后的数据 bostonStd_1所有行 前13列），y（标准化后的数据 bostonStd_1 所有行 第14列））为train,test两部分，
#train部分的比例是30%

print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

```python
#fit(训练)上面分好的训练集
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
#print("特征参数值：%s"%lr.coef_)
#print("截距：%s"% lr.intercept_)
#predict(预测)
y_pred = lr.predict(X_test)
lr.score(X_test,y_test)#【遗留问题：精度得分100%?】
```

```python
#评估模型

#【评估原理：比较y_test实际值，和上面求得的y_pred预测值】

from sklearn import metrics
#MSE评估指标 均方误差
skl_MSE = metrics.mean_squared_error(y_test,y_pred)

#RMSE评估指标 均方根误差
skl_RMSE = np.sqrt(skl_MSE)

#算MAE平均绝对值误差 #单词absolute，绝对的
skl_MAE = metrics.mean_absolute_error(y_test,y_pred)
#算R^2
skl_R2 = metrics.r2_score(y_test,y_pred)

print("skl算MSE: %s" % skl_MSE)
print("skl算RMSE: %s" % skl_RMSE)
print("skl算R方: %s" % skl_R2)
print("skl算MAE: %s" % skl_MAE)
```

```python
#【上面改近方向：可用Lasso回归梯度下降加入学习率训练，求得最优参数，和学习率之间的平衡】

#k近邻 网格搜索和GridSearchCV调参来实现多变量线性回归的找到最优参数，测算精确度，看看是不是使回归更优
from sklearn.neighbors import KNeighborsRegressor
from sklearn.grid_search import GridSearchCV

knn_reg = KNeighborsRegressor()

param_grid = [
    {
        'weights': ['uniform'],
        'n_neighbors':[i for i in range(1,11)]
    },
    {
        'weights':['distance'],
        'n_neighbors':[i for i in range(1,11)],
        'p':[i for i in range(1,6)]
    }
]

knn_reg = KNeighborsRegressor()

grid_search=GridSearchCV(knn_reg,param_grid,n_jobs=-1,verbose=2)
grid_search.fit(X_train,y_train)

#最优参数
grid_search.best_params_
#{'n_neighbors': 2, 'p': 1, 'weights': 'distance'}

#代入最优参数得到算法精确度
grid_search.best_estimator_.score(X_test,y_test)
#98.73%

#对比上面LinearRegression看下，网格搜索K近邻 多元线性回归求得的精确度为98%，
#LinearRegression没用学习率和参数调优。100%。可能会产生过拟合。
```
